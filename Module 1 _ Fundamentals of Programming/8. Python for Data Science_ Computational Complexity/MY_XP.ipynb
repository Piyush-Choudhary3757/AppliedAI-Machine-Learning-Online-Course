```python
import math
def binarySearch(arry,l,r,x):

    #Check Base Case
    if r >=l:
        mid = l + math.floor((r-l)/2)

        #if Element is present at the middle itself
        if arry[mid]== x:
            return mid
        
        #if element is smaller than mid ,then it can only be present in the left subarray
        elif arry[mid] > x:
            return binarySearch(arry,l,mid-1,x)
        
        #else the element can only be present in the right subarray
        else: 
            return binarySearch(arry,mid+1,r,x)
    else:
        #element is not present in the array
        return -1
    
l.sort();
arr = l;
q = 31;
binarySearch(arry,0,len(arr)-1,q)
```

```python
# Find element common in two lists:
l1 = list(range(100))
random.shuffle(l1)

l2 = list(range(50))
random.shuffle(l2)

#find common elements : O(n*m)

cnt=0;
for i in l1:
    for j in l2:
        if i==j:
            print(i)
            cnt += 1;
print("Number of common elements:", cnt)
```

```python
l1 = list(range(100))
random.shuffle(l1)
 
l2 = list(range(50))
random.shuffle(l2)

#find common elements in list in O(n) time and O(m) space if m<n

##add all elements in the smallest list into a hashtable/Dict: O(m) space

smallList = {}
for ele in l2:
    smallList[ele] = 1; #any value is ok .Key is important

#now find common elements

cnt=0;
for i in l1:
    if smallList.get(i) != None: #search happens in constant time.
        print(i);
        cnt += 1;
print("number of common elements :",cnt)
```
