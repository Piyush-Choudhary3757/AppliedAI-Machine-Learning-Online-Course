```python
#PDF
haberman_survived = hbs[hbs["status"] == 1];
haberman_notsurvived = hbs[hbs["status"] == 2];

#print(haberman_survived)
#print(haberman_notsurvived)

plt.plot(hbs["age"], np.zeros_like(hbs['age']), 'o')
plt.plot(hbs["year"], np.zeros_like(hbs['year']), 'o')

plt.xlabel("nodes");
plt.ylabel("status");
plt.legend();
plt.title("Survival vs nodes graph");

plt.show()
    
```

```python

sns.FacetGrid(hbs, hue="status", size=5) \
   .map(sns.distplot, "status") \
   .add_legend();
plt.show();
```

```python
#CDF
counts, bin_edges = np.histogram(hbs['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)



plt.show();
```

```python
hbs.plot(kind="scatter", x = 'age', y = 'year', title= 'Year vs Age');
plt.show()
```
